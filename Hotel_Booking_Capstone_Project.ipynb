```python
set_path = '/content/drive/MyDrive/AlmaBetter/Colab Notebooks/Copy of Hotel Bookings.csv'
```

```python
df_ht = pd.read_csv(set_path)
```

```python
# Counting the total null values in each columns
count = df_ht.isnull().sum()
#Percentage of total null values in each columns
percentage = round(count/df_ht.isnull().count(),4) *100
#concating the count and percentage
df_temp = pd.concat([count,percentage], axis = 1, keys = ['null_count', 'null_percent'])
```

```python
#Droping the columns with null values greater than 15 percent
df_ht = df_ht.drop((df_temp[df_temp['null_percent'] > 15]).index,1)
```

```python
# Function to convert month_name to month_number
def num_month(month_name):
  if month_name == 'January':
    month_name = '01'
  elif month_name == 'February':
    month_name = '02'
  elif month_name == 'March':
    month_name = '03'
  elif month_name == 'April':
    month_name = '04'
  elif month_name == 'May':
    month_name = '05'
  elif month_name == 'June':
    month_name = '06'
  elif month_name == 'July':
    month_name = '07'
  elif month_name == 'August':
    month_name = '08'
  elif month_name == 'September':
    month_name = '09'
  elif month_name == 'October':
    month_name = '10'
  elif month_name == 'November':
    month_name = '11'
  elif month_name == 'December':
    month_name = '12'
  return month_name
#Function to convert single_digit_date to double_digit_date
def date_two_digit(single_digit_date):
  if single_digit_date == 1:
    single_digit_date = '01'
  elif single_digit_date == 2:
    single_digit_date = '02'
  elif single_digit_date == 3:
    single_digit_date = '03'
  elif single_digit_date == 4:
    single_digit_date = '04'
  elif single_digit_date == 5:
    single_digit_date = '05'
  elif single_digit_date == 6:
    single_digit_date = '06'
  elif single_digit_date == 7:
    single_digit_date = '07'
  elif single_digit_date == 8:
    single_digit_date = '08'
  elif single_digit_date == 9:
    single_digit_date = '09'
  elif single_digit_date == 10:
    single_digit_date = '10'
  return single_digit_date
```

```python
#Applying num_month function to df
df_ht['month_num'] = df_ht['arrival_date_month'].apply(num_month)
#Applying date_two_digit function to df
df_ht['arrival_date_day_of_month'] = df_ht['arrival_date_day_of_month'].apply(date_two_digit)
#Assigning month_num to arrival_date_month
df_ht['arrival_date_month'] = df_ht['month_num']
#droping month_num
df_ht.drop(['month_num'], axis = 1, inplace = True)
```

```python
#Concating columns to get date in str format
df_ht['date_'] = df_ht['arrival_date_year'].astype(str)+ '-' + df_ht['arrival_date_month'].astype(str) + '-' + df_ht['arrival_date_day_of_month'].astype(str) 
```

```python
#Converting the date from str to DATE format 
df_ht['date'] = pd.to_datetime(df_ht.date_)
```

```python
#droping temporary column
df_ht.drop(['date_'], axis = 1)
```
